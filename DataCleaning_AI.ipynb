```python
import pandas as pd

df = pd.read_csv("Online_Courses_Final_4000_Updated.csv")

print("Total Rows (Raw):", len(df))
print("Unique Domains:", df["domain"].unique())
print("Unique Levels:", df["level"].unique())
```

```python
df["domain"] = df["domain"].str.strip()

# Standardize variations
df["domain"] = df["domain"].replace({
    "Gen Ai": "GenAI",
    "Generative AI": "GenAI",
    "genai": "GenAI",
    "Devops": "DevOps",
    "QA Testing": "QA"
})
```

```python
print("Unique Domains After Clean:", df["domain"].unique())
```

```python
df["level"] = df["level"].str.strip().str.title()

df["level"] = df["level"].replace({
    "Easy": "Beginner",
    "Medium": "Intermediate",
    "Hard": "Advanced"
})
```

```python
print("Unique Levels After Clean:", df["level"].unique())
```

```python
difficulty_map = {
    "Beginner": 1,
    "Intermediate": 3,
    "Advanced": 5
}

df["difficulty_score"] = df["level"].map(difficulty_map)
df["difficulty_tag"] = df["level"]
```

```python
df["recommended_for_score_range"] = df["recommended_for_score_range"].astype(str).str.replace(" ", "")

df[["score_min", "score_max"]] = df["recommended_for_score_range"].str.split("-", expand=True)

df["score_min"] = pd.to_numeric(df["score_min"], errors="coerce")
df["score_max"] = pd.to_numeric(df["score_max"], errors="coerce")
```

```python
print(df[["score_min", "score_max"]].head())
```

```python
df["skills_covered"] = (
    df["skills_covered"]
    .fillna("")
    .str.lower()
)
```

```python
df = df.dropna(subset=["course_id", "domain", "level", "score_min", "score_max"])
```

```python
print("Total Rows (Raw):", len(df))
```

```python
df = df.drop_duplicates(subset="course_id")
```

```python
print("Final Row Count:", len(df))

print("\nDomain Distribution:")
print(df["domain"].value_counts())

print("\nLevel Distribution (All):")
print(df["level"].value_counts())

print("\nGenAI Levels:")
print(df[df["domain"] == "GenAI"]["level"].value_counts())

print("\nDuplicate IDs:", df["course_id"].duplicated().sum())
```

```python
final_columns = [
    "course_id",
    "course_title",
    "domain",
    "level",
    "difficulty_score",
    "difficulty_tag",
    "estimated_duration_hours",
    "xp_reward",
    "recommended_for_mastery",
    "primary_skill",
    "skills_covered",
    "subtopics",
    "tools_used",
    "prerequisites",
    "course_description",
    "learning_outcomes",
    "rating",
    "url",
    "score_min",
    "score_max"
]

df = df[final_columns]
```

```python
df.to_csv("Online_Courses_Cleaned_Final.csv", index=False)
```
